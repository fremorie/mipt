```python
import matplotlib.pyplot as plt

%matplotlib inline
```

Pseudorandom numbers may be generated by using the following formula:

$$z_{n+1} = \mu z_{n} + \beta\; (\mathrm{mod}\; p) $$


```python
import numpy as np

### function to generate random numbers (by formula above)

def pseudorandom(z0, mu, p, beta, number_of_numbers):
    zs = [z0]
    zn = z0
    for i in range(number_of_numbers):
        zn1 = (mu*zn + beta)%p
        zs.append(zn1)
        zn = zn1
    return np.array(zs)/float(p)
    
```

Let's play with the parameters mu, beta, p.

```python
### beta = 0, mu very small

random = pseudorandom(z0 = 1l, mu = 7, p = 104729, beta = 0, number_of_numbers = 1000)
```

```python
plt.scatter(random[:-1], random[1:])
```

```python
# beta = 0, mu = 10

random = pseudorandom(z0 = 1l, mu = 10, p = 104729, beta = 0, number_of_numbers = 1000)
plt.scatter(random[:-1], random[1:])
```

```python
### beta = 0, mu = 30

random = pseudorandom(z0 = 1l, mu = 30, p = 104729, beta = 0, number_of_numbers = 1000)
plt.scatter(random[:-1], random[1:])
```

```python
### beta = 100, mu = 30

random = pseudorandom(z0 = 1l, mu = 30, p = 104729, beta = 18974, number_of_numbers = 1000)
plt.scatter(random[:-1], random[1:])
```

```python
### beta = 0, mu is quite big

random = pseudorandom(z0 = 1l, mu = 19897, p = 104729, beta = 0, number_of_numbers = 1000)
plt.scatter(random[:-1], random[1:])
```
