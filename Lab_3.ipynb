### Lab 3: Виннеровский процесс (броуновское движение по X и Y)

Импортируем необходимые библиотеки:

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

Задаем количество шагов n, шаг по времени dt и сами шаги: 

```python
n = 100
dt = 1.0

delta_X = np.random.standard_normal(size=(n, 2)) * dt
```

Задаем координаты (x,y) в матрице X:

```python
X = np.cumsum(delta_X, axis = 1)
```

```python
plt.plot(X[:, 0], X[:, 1])
```

Зависимости x(t) и y(t):

```python
plt.plot(X[:,0])
```
